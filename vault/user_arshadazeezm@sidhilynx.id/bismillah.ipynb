> **1. q1.c**: Write a function that takes an integer and returns a dictionary of its prime factors and their exponents. Add a test case for the number 315.

```python
def prime_factorization(n):
    """
    Calculates the prime factorization of a given integer.

    Args:
        n: An integer.

    Returns:
        A dictionary where keys are prime factors and values are their exponents.
    """
    factors = {}
    d = 2
    temp = n
    while d * d <= temp:
        while temp % d == 0:
            factors[d] = factors.get(d, 0) + 1
            temp //= d
        d += 1
    if temp > 1:
        factors[temp] = factors.get(temp, 0) + 1
    return factors

# Test case
number_to_factorize = 315
result = prime_factorization(number_to_factorize)
print(f"The prime factorization of {number_to_factorize} is: {result}")
```

> **2. q2.c**: Given a list of dictionaries containing student names and their scores across three subjects, calculate the average for each student and identify the "Top Performer."

```python
def calculate_student_averages(students_data):
    """
    Calculates the average score for each student and identifies the top performer.

    Args:
        students_data: A list of dictionaries, where each dictionary represents a
                       student and contains 'name' (str) and 'scores' (list of int/float).

    Returns:
        A tuple containing:
        - A dictionary of student names and their calculated averages.
        - The name of the top-performing student (str).
        - The average score of the top-performing student (float).
    """
    student_averages = {}
    top_performer_name = ""
    top_performer_average = -1.0

    for student in students_data:
        name = student['name']
        scores = student['scores']
        if scores:  # Ensure there are scores to calculate an average
            average = sum(scores) / len(scores)
            student_averages[name] = average

            if average > top_performer_average:
                top_performer_average = average
                top_performer_name = name
        else:
            student_averages[name] = 0.0 # Assign 0 if no scores

    return student_averages, top_performer_name, top_performer_average

if __name__ == "__main__":
    # Sample student data
    students_list = [
        {"name": "Alice", "scores": [85, 90, 78]},
        {"name": "Bob", "scores": [92, 88, 95]},
        {"name": "Charlie", "scores": [76, 82, 79]},
        {"name": "Diana", "scores": [95, 91, 93]}
    ]

    averages, top_performer, top_avg = calculate_student_averages(students_list)

    print("--- Student Averages ---")
    for student, avg in averages.items():
        print(f"{student}: {avg:.2f}")

    print("\n--- Top Performer ---")
    print(f"The top performer is: {top_performer}")
    print(f"Their average score is: {top_avg:.2f}")
```

> **3. q3.c**: Create a 3x3 matrix using nested lists. Write a list comprehension that transposes the matrix and prints the result in a formatted grid.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transposed_matrix = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

# Print the transposed matrix in a formatted grid
print("Transposed Matrix:")
for row in transposed_matrix:
    print(" ".join(map(str, row)))
```

> **4. q4.c**: Implement a function that uses a closure to return the next number in the Fibonacci sequence every time it is called. Generate and print the first 10 numbers.

```python
def fibonacci_generator():
    a, b = 0, 1
    def next_fib():
        nonlocal a, b
        result = a
        a, b = b, a + b
        return result
    return next_fib

# Generate and print the first 10 Fibonacci numbers
fib_gen = fibonacci_generator()
print("Generating the first 10 Fibonacci numbers:")
for _ in range(10):
    print(fib_gen())
```
