> 4.Write a program to find the sum, difference, product, and quotient of two numbers.

```python
print("Hello! Let's do some math!")

number1_str = input("Please enter the first number: ")
number2_str = input("Please enter the second number: ")

number1 = float(number1_str)
number2 = float(number2_str)

sum_result = number1 + number2
difference_result = number1 - number2
product_result = number1 * number2

print("The sum of", number1, "and", number2, "is:", sum_result)
print("The difference between", number1, "and", number2, "is:", difference_result)
print("The product of", number1, "and", number2, "is:", product_result)

if number2 != 0:
    quotient_result = number1 / number2
    print("The quotient of", number1, "divided by", number2, "is:", quotient_result)
else:
    print("Cannot divide by zero!")

print("Thanks for using the calculator!")
```

> 6.Write a program to calculate the area of a circle (πr²).

```python
import math

def calculate_circle_area(radius):
    area = math.pi * radius * radius
    return area

user_radius_str = input("Enter the radius of the circle: ")
user_radius = float(user_radius_str)

calculated_area = calculate_circle_area(user_radius)

print("The area of the circle is:", calculated_area)
```

> 7.Write a program to check whether a number is even or odd.

> 11.Write a program to find the sum of first 10 natural numbers.

```python
total_sum = 0
number = 1

while number <= 10:
    total_sum = total_sum + number
    number = number + 1

print("The sum of the first 10 natural numbers is:", total_sum)
```

> 12.Write a program to print the multiplication table of a given number.

```python
def print_multiplication_table(number):
    print("Multiplication Table for", number)
    for i in range(1, 11):
        result = number * i
        print(number, "x", i, "=", result)

user_input_number = int(input("Enter a number to see its multiplication table: "))
print_multiplication_table(user_input_number)
```

> 13.Write a program to swap two numbers (without using a third variable).

```python
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

print("Before swapping:")
print("First number is:", number1)
print("Second number is:", number2)

number1 = number1 + number2
number2 = number1 - number2
number1 = number1 - number2

print("After swapping:")
print("First number is:", number1)
print("Second number is:", number2)
```

> 14.Write a program to find the factorial of a number.

```python
def calculate_factorial(num):
    if num < 0:
        return "Factorial is not defined for negative numbers"
    elif num == 0:
        return 1
    else:
        result = 1
        for i in range(1, num + 1):
            result = result * i
        return result

user_input_number = int(input("Enter a number to find its factorial: "))

final_answer = calculate_factorial(user_input_number)

print(f"The factorial of {user_input_number} is {final_answer}")
```

> 15.Write a program to reverse a given number.

```python
number = input("Enter a number: ")

reversed_number = ""

for digit in number:
    reversed_number = digit + reversed_number

print("The reversed number is:", reversed_number)
```
